```python
def addSum(a, b, c):
    rst = a+b+c
    return rst

t = [10, 20, 30]
print(addSum(*t))
print(addSum([1, 2], [3, 4], [5, 6, 7]))
```

```python
def addSum(*a):
    rst = 0
    for i in a:
        rst += i
    return rst

t = [10, 20, 30]
print(addSum(*t))
print(addSum(1, 2, 3, 4))
```

```python
class Test:
    def __init__(self, a):
        self.a = a;
        
    def addSum(self, b):
        print(self.a + b)
        
class SubTest(Test):
    def sqr(self, c):
        print(c * c)
        
rv = SubTest(10)
rv.sqr(10)
rv.addSum(20)
```
